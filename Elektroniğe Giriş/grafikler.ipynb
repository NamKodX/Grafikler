# Kütüphane import etme bölümü

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

## Elektronik Deneyleri

### 1- RC ve RL Süzgeç Devreleri

```python
rc = pd.read_excel("rc.xlsx")  #RC süzgeç devresinin verilerini içeri aldık
```

```python
rc_data = pd.DataFrame(rc)     #RC süzgeç devresinin verilerini tablo haline getirdik
```

```python
rc_data                        #RC süzgeç devresinin bulunduğu tabloyu ekrana yazdırdık
```

```python
rc_data.info()       #RC verisi hakkında bilgiler 
```

```python
rl = pd.read_excel("rl.xlsx")  #RL süzgeç devresinin verilerini içeri aldık
```

```python
rl_data = pd.DataFrame(rl)     #RL süzgeç devresinin verilerini tablo haline getirdik
```

```python
rl_data                        #RL süzgeç devresinin bulunduğu tabloyu ekrana yazdırdık
```

```python
rl_data.info()          #RL verisi hakkında bilgiler
```

```python
rl_rc = pd.concat([rc_data,rl_data],axis=1) 
#RC ve RL veri tablosunu bir tabloda birleştirebilmek için kullanılır
```

```python
rl_rc              #İstersek RC ve RL veri tablosunu bir tabloda birleştirebiliriz
```

```python
#Grafikleri figür haline getirip ardından boyutlandırıyoruz
plt.figure(figsize=(20,10))
#Yüksek Frekanslı Verileri çizgi grafiği olarak çiziyoruz
rc_seri = plt.plot(rc_data["Frekans"],rc_data["Seri"],".-",label="Yüksek Frekans")
#Alçak Frekanslı Verileri çizgi grafiği olarak çiziyoruz
rc_paralel = plt.plot(rc_data["Frekans"],rc_data["Paralel"],".-",label="Alçak Frekans")
#Grafik başlığı yazıyoruz
plt.title("RC Süzgeç Grafiği")
#X eksenine isim veriyoruz
plt.xlabel("Frekans")
#Y eksenine isim veriyoruz
plt.ylabel("Veriler")
#Grafiği okumak kolaylaşsın diye ızgara çiziyoruz
plt.grid(True, color = "grey", linewidth = "1.4", linestyle = "-.") 
#Hangi çizginin neyi temsil ettiğini yazdırıyoruz
plt.legend()
```

```python
#Grafikleri figür haline getirip ardından boyutlandırıyoruz
plt.figure(figsize=(20,10))
#Yüksek Frekanslı Verileri çizgi grafiği olarak çiziyoruz
rl_seri = plt.plot(rl_data["Frekans"],rl_data["Seri"],".-",label="Yüksek Frekans")
#Alçak Frekanslı Verileri çizgi grafiği olarak çiziyoruz
rl_paralel = plt.plot(rl_data["Frekans"],rl_data["Paralel"],".-",label="Alçak Frekans")
#Grafik başlığı yazıyoruz
plt.title("RL Süzgeç Grafiği")
#X eksenine isim veriyoruz
plt.xlabel("Frekans")
#Y eksenine isim veriyoruz
plt.ylabel("Veriler")
#Grafiği okumak kolaylaşsın diye ızgara çiziyoruz
plt.grid(True, color = "grey", linewidth = "1.4", linestyle = "-.") 
#Hangi çizginin neyi temsil ettiğini yazdırıyoruz
plt.legend()
```

### 2 - Seri ve Paralel Rezonans devreler

```python
paralel_tablo                                     #Yapılan tabloyu ekrana yazdırıyoruz
```

```python
seri = pd.read_excel("seri_rezonans.xlsx")  #Seri rezonans grafiğin verilerini içeri aktarıyoruz
```

```python
#Paralel Rezonans grafiği çizdirme bölümü
#Grafikleri figür haline getirip ardından boyutlandırıyoruz
plt.figure(figsize=(20,10))
#Grafikteki x ve y bileşenleri belirleyip işleme koyuyoruz
paralel_grafik = plt.plot(paralel_tablo["frekans"],paralel_tablo["V0/Vi"],".-",color="red")
#Grafik başlığı yazıyoruz
plt.title("Paralel Rezonans Grafiği")
#X eksenine isim veriyoruz
plt.xlabel("Frekans")
#Y eksenine isim veriyoruz
plt.ylabel("V0/Vi")
#Grafiği okumak kolaylaşsın diye ızgara çiziyoruz
plt.grid(True, color = "grey", linewidth = "1.4", linestyle = "-.") 
plt.show()
```

```python
#Seri Rezonans devre çizdirme bölümü
#Grafikleri figür haline getirip ardından boyutlandırıyoruz
plt.figure(figsize=(20,10))
#Yüksek Frekanslı Verileri çizgi grafiği olarak çiziyoruz
seri_grafik = plt.plot(seri_tablo["frekans"],seri_tablo["V0/Vi"],".-",color="red")
#Grafik başlığı yazıyoruz
plt.title("Seri Rezonans Grafiği")
#X eksenine isim veriyoruz
plt.xlabel("Frekans")
#Y eksenine isim veriyoruz
plt.ylabel("V0/Vi")
#Grafiği okumak kolaylaşsın diye ızgara çiziyoruz
plt.grid(True, color = "grey", linewidth = "1.4", linestyle = "-.") 
plt.show()
```
